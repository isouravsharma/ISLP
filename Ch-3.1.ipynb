```python
import pandas as pd
import numpy as np
import math
import matplotlib.pyplot as plt
```

```python
rng = np.random.default_rng(seed = 1)
X = rng.random(100)
e = rng.standard_normal(100)
Y = 2 + 3*X + e
```

```python
fig, ax = plt.subplots(figsize=(15,6), nrows=1, ncols=3)
ax[0].plot(X)
ax[1].plot(Y)
ax[2].scatter(X,Y)
```

```python
fig, ax = plt.subplots(figsize=(15,5), nrows=1, ncols=1)
ax.scatter(X,Y, alpha= 0.7, s = 60, edgecolors="k")
b, a = np.polyfit(X, Y, deg=1)
ax.plot(X, a+b*X, lw = "2.5");
```
