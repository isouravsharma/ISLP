```python
rng = np.random.default_rng(seed = 1)
X = rng.standard_normal(10)
X 
```

```python
# SD
np.std(X), np.mean(X), np.std(X)/len(X)
```

```python
plt.plot(X)
plt.axhline(np.mean(X), color = 'red'); # Draw horizontal line of Mean
```

```python
# Confidence Interval 

np.mean(X) + (1.9600*(np.std(X)/len(X))), np.mean(X) - (1.9600*(np.std(X)/len(X)))
```

```python
plt.plot(X)
plt.axhline(np.mean(X), color = 'red'); # Draw horizontal line of Mean
plt.axhline(np.mean(X) + 2*(1.036*(np.std(X)/len(X))))
plt.axhline(np.mean(X) - 2*(1.9600*(np.std(X)/len(X))))
```
